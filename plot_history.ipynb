```python
from pickle import load

file = open("./study_info.pkl", "rb")
study_info = load(file)
```

```python
import numpy as np
print(len(study_info['history']['f1_loss_valid']))
print(study_info['history'].keys())
print(np.min(study_info['history']['f1_loss_train']))
print(np.min(study_info['history']['f1_loss_valid']))
```
